## Creating a continuum map
Next, we are going to create a continuum map of the galaxy. For this we want to do couple of things first:
1. We will mask all of the emission lines. This will ensure that the continuum map is only continuum and no emission lines
2. We will do extremely rudementary sky line residual masking (we will talk about it later)/
3. We will then collapse this cube in wavelength direction by taking a median value of each spectrum - median continuum map

```python
Cube.mask_emission()
Cube.mask_sky( 1.5)
Cube.collapse_white(plot_it)
plt.show()
```

## Where is my galaxy?
So, lets find the galaxy itself. The galaxy deosnt have to be necessarily in the center of the frame. The guiding is not perfect and can shift. The Find center function fits a 2D gaussian to the continuum image to find the center. If there is a problem with the fitting (continuum not detected, bad noise on the edges, etc.) we can manually input the center of the galaxy. 

After that we will "choose pixels". This essentially selects spaxels within a circular aperture within a defined radius. I usually start with radius of 0.3 arcsecond - diameter of 0.6 arcseconds - typical seeing (resolution) with KMOS. With JWST, who knows we will test things.

```python
Cube.find_center( plot_it) #manual = [20,20]
Cube.choose_pixels(1, rad= 0.3)
plt.show()
```

## Stacking the sky spectrum and masking skyline residuals

As you might have heard during the MOONS meetings, the sky is extremely noisie in the NIR. It is full of emission lines emitted by the molecules in the air. However, we know where they are and we can model the sky spectrum, or take a sky frame (exposure of the same time of blank sky without any objects in it) and subtract it from our science frame. 

Whatever method you decide to use, there will always be some residuals left. The spectrograph flexes (and other issues) so that the skyline will appear on a slighlty different channel (spectral pixel) and hence there will be a huge spike left. 

In order to deal with it, we have implemented a little algorithm. We take all of the spaxels from that do not include the objects (the outer edges) and sum these spectra up for a master sky spectrum. As we sum the spectra the noise goes down (as it is random around 0), while the skyline residuals go up because they are not random. We then mask everything that is over 2 sigma above the noise. This will hopefully cut out all of the skyline residuals. 

```python
Cube.stack_sky(plot_it)
plt.show()
```

Detecting an outflow in Halpha is pretty hard and it will not be a common favoured model. Furthermore, the fits with the outflow are quite degenerate and any results should be viewd with caution.

## Creating maps:

It is now to take advantage of the full Integral field spectroscopy part of the data. Since every spatial pixel (spaxel) contains a spectrum we can map the emission line across the objects. In order to fit each spaxel and use the multple cores on our machine, I first unpack the cube into the individual spectra. 

We define the size of the aperture which we would like to map - no reason to fit every spaxel in the cube. 

We also have an option to bin the spaxel together in order to boost the SNR. Since the spatial resolution of KMOS is 0.4-0.6 arcseconds and the spatial scale is 0.2 arcseconds (unless rebinned to 0.1 arcsecond scale), we can take a median spectrum of the spaxel and the neighboring spaxel around it. This boosts the signal significantly without reducing the spatial scale in KMOS and seeing mode of SINFONI. Lets see what happens with JWST. If you dont want any binning, use keyword 'Individual'.

```python
Cube.unwrap_cube(rad=1.1, sp_binning='Nearest')
```
